```python
import glob
import pandas as pd
```

```python
path=r'./한국소비자원 생필품 가격데이터 2019년 2분기'
datas=glob.glob(path+'/*.csv')
datas
```

```python
datas2=[]
for data in datas:
    df=pd.read_csv(data,encoding='euc-kr')
    print('이름:{} - 크기:{}'.format(data,df.shape))
    datas2.append(df)

price_data= pd.concat(datas2, ignore_index=True)
price_data.head()
```
