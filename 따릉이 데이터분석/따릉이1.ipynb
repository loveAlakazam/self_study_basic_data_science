## df2018_3tmp: class str -> int / float 타입으로 변경 

### to_numeric()을 사용한다.

(참고자료 url) https://code-examples.net/ko/q/f27a5e

```python
for col, col_tmp in zip(df2018_3.columns, df2018_3tmp.columns):
    original_type=df2018_3[col].dtypes
    tmp_type= df2018_3tmp[col_tmp].dtypes
    
    if original_type !=tmp_type:
        # original_type는 int64,float64인데, tmp_type은 str이라면
        df2018_3tmp[col_tmp]=df2018_3tmp[col_tmp].apply(pd.to_numeric)

df2018_3tmp
```

```python
for col in df2018_3tmp.columns:
    col_type=df2018_3tmp[col].dtypes
    print(col_type)
```

# df2018_3tmp와 df2018_3을 연결하기 

```python
df2018_3= pd.concat([df2018_3tmp, df2018_3], ignore_index=True)
df2018_3.head()
```
