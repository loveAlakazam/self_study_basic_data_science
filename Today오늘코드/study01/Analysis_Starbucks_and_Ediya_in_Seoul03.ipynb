# 스타벅스 이디야 위치 분석

```python
import glob
import pandas as pd
import warnings
# warning message 숨기기
warnings.filterwarnings(action='ignore')

# 2019년도 6월 데이터 
path=r'./datas/06'
stores_datas= glob.glob(path+'/*.csv')

concat_csv_files=[]
for csv_file in stores_datas:
    df= pd.read_csv(csv_file, encoding='utf-8')
    print('file-name: {}\nsize: {}'.format(csv_file, df.shape))
    concat_csv_files.append(df)

store_datas= pd.concat(concat_csv_files, ignore_index=True)
store_datas.head(3)
```

### 스타벅스 데이터 수집 (전국단위)

```python
starbucks= store_datas[ store_datas['상호명'].str.contains('스타벅스|starbucks|STARBUCKS', na=False)]
starbucks.shape
```

```python
# ediya데이터와 starbucks 데이터 합치기 (동일한 컬럼)
# df_cafe= pd.concat([starbucks, ediya], ignore_index=True)
```

```python
# df_cafe_all는 전국의 이디야와 스타벅스 인 상점데이터를 저장.
df_cafe_all= store_datas[ store_datas['상호명'].str.contains('이디야|EDIYA|ediya|스타벅스|starbucks|STARBUCKS', na=False)]
df_cafe_all.head(5)
```

<br><br>
## 서울 특별시 - 이디야와 스타벅스 위치 분석

```python
# 시도명을 기준으로 분류
df_cafe_all['시도명'].value_counts()
```

```python
#  '시도명'이 서울특별시인 데이터를 수집하겠다.
#  특정컬럼만 선택
df_cafe_seoul= df_cafe_all[ df_cafe_all['시도명']=='서울특별시'][['상호명','시도명','시군구명','지번주소', '도로명주소', '위도', '경도']]
df_cafe_seoul.head(5)
```

<br>

### 브랜드명 컬럼 생성하기 

```python
df_cafe_seoul.loc[ df_cafe_seoul['상호명'].str.contains('스타벅스|starbucks|STARBUCKS'), '브랜드명']='스타벅스'
df_cafe_seoul.head(5)
```

```python
# 위의 결과에 따르면 '이디야'는 브랜드명 NaN으로 되어있다.
# NaN인것을 찾아서 브랜드명을 '이디야'로 바꾼다.

#  '브랜드명' 컬럼에서 fillna()이용하여 NaN 대신 '이디야'로 채운다.
df_cafe_seoul['브랜드명']= df_cafe_seoul['브랜드명'].fillna('이디야')
df_cafe_seoul.head(5)
```

```python
# isnull 실습
df_cafe_seoul['브랜드명'].isnull()
```

```python
# 브랜드명이  NaN인 데이터만 나타내기
df_cafe_seoul.loc[ df_cafe_seoul['브랜드명'].isnull()]
```

```python
df_cafe_seoul['브랜드명'].value_counts()
```

# 이디야와 스타벅스 지도에 찍어보기

```python
import folium

geo_df= df_cafe_seoul
# 지도를 보여주기 위한 초기의 중심위치
map = folium.Map(location=[geo_df['위도'].mean(),
                           geo_df['경도'].mean()],
                 zoom_start=12)

for n in geo_df.index:
    #팝업에 들어갈 텍스트를 지정한다.
    popup_name= geo_df.loc[n,'상호명']+ ' - '+ geo_df.loc[n,'지번주소']

    # 브랜드명에 따라 아이콘 색상을 달리하여 찍는다.
    if geo_df.loc[n,'브랜드명']=='스타벅스':
        icon_color='green'
    else:#브랜드명: 이디야
        icon_color='blue'

    folium.CircleMarker(
        location=[geo_df.loc[n,'위도'], geo_df.loc[n,'경도']],
        radius=3,
        popup=popup_name,
        color=icon_color,
        fill=True,
        fill_color=icon_color
    ).add_to(map)
map
```

# 서울특별시 스타벅스 매장분포

```python
starbucks= df_cafe_seoul[df_cafe_seoul['브랜드명']=='스타벅스']
starbucks
```

```python
starbucks_cnt=pd.DataFrame(
    df_cafe_seoul[df_cafe_seoul['브랜드명']=='스타벅스'].groupby('시군구명')['상호명'].count()
)
starbucks_cnt.columns=['매장수']
starbucks_cnt
```

```python
import json
geo_path = './datas/02. skorea_municipalities_geo_simple.json'
geo_json = json.load( open(geo_path, encoding='utf-8'))

geo_df= starbucks
# 지도를 보여주기 위한 초기의 중심위치 설정: 서울시의 스타벅스 위도/경도의 평균값으로 중심점을 설정.
map = folium.Map(location=[geo_df['위도'].mean(),
                           geo_df['경도'].mean()],
                 zoom_start=11, tiles='Stamen Toner')
map
```

```python
import json
geo_path = './datas/02. skorea_municipalities_geo_simple.json'
geo_json = json.load( open(geo_path, encoding='utf-8'))

geo_df= starbucks
# 지도를 보여주기 위한 초기의 중심위치
map = folium.Map(location=[geo_df['위도'].mean(),
                           geo_df['경도'].mean()],
                 zoom_start=11, tiles='Stamen Toner')


# 스타벅스의 개수를 색깔 영역으로 표시
map.choropleth(geo_data=geo_json,
               data=starbucks_cnt['매장수'],
               columns=[starbucks_cnt.index, starbucks_cnt['매장수']],
               key_on='feature.properties.name',
               fill_color='YlGn',
               fill_opacity=0.6,
               line_opacity=0.2
              )


# 동그라미 마커로 스타벅스 위치 표시
geo_df=starbucks
for n in geo_df.index:
    #팝업에 들어갈 텍스트를 지정한다.
    popup_name= '(도로명주소): '+ geo_df.loc[n,'도로명주소']

    
    #if geo_df.loc[n,'브랜드명']=='스타벅스':
    icon_color='green'

    folium.CircleMarker(
        location=[geo_df.loc[n,'위도'], geo_df.loc[n,'경도']],
        radius=3,
        tooltip=popup_name,
        color=icon_color,
        fill=True,
        fill_color=icon_color
    ).add_to(map)
map
```

# 서울특별시 이디야 매장분포 

```python
ediya=df_cafe_seoul.loc[ df_cafe_seoul['브랜드명']=='이디야']
ediya
```

```python
ediya_cnt= pd.DataFrame(
    ediya.groupby(['시군구명'])['상호명'].count()
)

ediya_cnt.columns=['매장수']

ediya_cnt
```

```python
import json
geo_path = './datas/02. skorea_municipalities_geo_simple.json'
geo_json = json.load( open(geo_path, encoding='utf-8'))

geo_df= ediya
# 지도를 보여주기 위한 초기의 중심위치
map = folium.Map(location=[geo_df['위도'].mean(),
                           geo_df['경도'].mean()],
                 zoom_start=11, tiles='Stamen Toner')


# 이디야가게 개수를 색깔 영역으로 표시
geo_df=ediya_cnt
map.choropleth(geo_data=geo_json,
               data=geo_df['매장수'],
               columns=[geo_df.index, geo_df['매장수']],
               key_on='feature.properties.name',
               fill_color='PuBu',
               fill_opacity=0.6,
               line_opacity=0.2
              )


# 동그라미 마커로 이디야 위치 표시
geo_df=ediya
for n in geo_df.index:
    #팝업에 들어갈 텍스트를 지정한다.
    popup_name= '(주소) '+ geo_df.loc[n,'지번주소']

    icon_color='blue'

    folium.CircleMarker(
        location=[geo_df.loc[n,'위도'], geo_df.loc[n,'경도']],
        radius=3,
        tooltip=popup_name,
        color=icon_color,
        fill=True,
        fill_color=icon_color
    ).add_to(map)
map
```
