# 네이버 영화 평점 기준 영화의 평점 변화 확인하기 

[url] https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&date=20190903


```python
from bs4 import BeautifulSoup
import pandas as pd
from urllib.request import urlopen

url_base='http://movie.naver.com/'
url_sub='movie/sdb/rank/rmovie.nhn?sel=cur&date=20190903'

page= url_base+ url_sub
soup= BeautifulSoup(urlopen(page))
soup
```

```python
# 현재 페이지의 코드중
# class 이름이 'tit5'인 div 태그들을 모두 찾는다.
soup.find_all('div', 'tit5') 
```

```python
# tit5 클래스이름의 div태그중 0번째이며, a태그의 텍스트를 추출
soup.find_all('div', 'tit5')[0].a.get_text()
# 다른표현: soup.find_all('div', 'tit5')[0].a.string
```

```python
soup.find_all('div', 'tit5')[1].a.string
```

```python
# 점수 찾기
# 클래스이름이 point인 td 태그이다.
point=[]
movie_scores=soup.find_all('td', 'point')
for score in movie_scores:
    point.append(score.string)
point
```

```python
dates= pd.date_range('2019-5-25', periods=100, freq='D' )
dates
```

```python
import urllib
from tqdm import tqdm_notebook

movie_date=[]
movie_name=[]
movie_point=[]

# 100일동안 영화순위를 확인하기.
for today in tqdm_notebook(dates):
    html='https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&tg=0&date={date}'
    response= urlopen(html.format( date= urllib.parse.quote(today.strftime('%Y%m%d'))))
    soup= BeautifulSoup(response, 'html.parser')
    
    # class이름이 point인 td태그를 찾는다.
    end= len(soup.find_all('td','point')) #랭크의 개수를 구한다.
    
    movie_date.extend([ today for n in range(0,end)]) #리스트의 원소들(영화순위)만을 추가.
    movie_name.extend([ soup.find_all('div', 'tit5')[n].a.string for n in range(0,end)])
    movie_point.extend([ soup.find_all('td', 'point')[n].string for n in range(0,end)])
```

```python
movie= pd.DataFrame({'date': movie_date, 'name':movie_name, 'point':movie_point})
movie.head()
```

```python
# 책을 그대로 따라했는데, pivot_table 에서 aggfunc이 np.sum일때 
# 결과인 point부분이 문자열을 이어붙이는 꼴..

import numpy as np
# name을 기준으로 인덱스, point를 합친다.
movie_unique= pd.pivot_table(movie, index=['name'], values=['point'], aggfunc=np.sum)
movie_unique.head()
```

```python
# type이 string이라서 np.sum 연산의 결과가 문자열을 이어붙임.
movie['point'].types 
```

```python
# bs4.element.NavigableString => float 으로 변환
# 타입변경 astype
movie['point']=movie['point'].astype(np.float32)

# name을 기준으로 인덱스, point를 합친다.
movie_unique= pd.pivot_table(movie, index=['name'], values=['point'], aggfunc=np.sum)
movie_unique.head()
```

```python
movie_best = movie_unique.sort_values( by='point', ascending=False)
movie_best.head()
```
