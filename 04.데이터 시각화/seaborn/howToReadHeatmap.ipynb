```python
import numpy as np
np.random.seed(0)

import seaborn as sns
sns.set()

uniform_data = np.random.rand(10, 12)
ax= sns.heatmap(uniform_data)
```

```python
# 컬러맵의 limit를 바꾼다. (오른쪽 확률 컬러 막대기(ax)의 최소와 최대부분을 말함)
ax= sns.heatmap( uniform_data, vmin=0, vmax=1)
# 값이 클수록 색깔이 연하다.
# 값이 작을수록 색깔이 진하다.
```

```python
# 0을 가운데로 넣는다 (컬러맵이니까 0은 검정색?)
# 0을 가운데로해서 컬러맵을 나눈다.

normal_data= np.random.randn(10, 12)
ax= sns.heatmap( normal_data, center=0 )

# 음수값은 파란색
# 0에 가까울수록 검정색
# 양수값은 빨강색
```

```python
# 데이터 프레임을 row와 color 라벨에 의미가 있는...
flights = sns.load_dataset("flights")
flights = flights.pivot( "month", "year", "passengers")
ax= sns.heatmap(flights)

# month : y축
# year: x 축
# passengers 셀에 있는색깔을 값으로 나타냄..
# passengers 수가 적을수록 색깔이 어둡고, 많을수록 색깔이 밝음..
```

```python
# 히트맵의 셀에 값을 표시하기
ax= sns.heatmap( flights, annot=True, fmt="d")
```

```python
# fmt가 없다면?
ax= sns.heatmap( flights, annot=True)
# fmt없으면 크기 셀안에 데이터값이 들어있지않고 엉성하게 만듦...
```

```python
# 셀을 구분시켜놓자
ax= sns.heatmap( flights, linewidths=.5, annot=True, fmt="d")
```

```python
# 히트맵 색깔 바꿔보기
ax=sns.heatmap( flights, cmap="YlGnBu")
```

```python
# 특정한 값을 중앙으로하는 히트맵그리기
# center 좌표가 [ "January", 1955] 이다. center을 중앙점(0=검정색)으로 함..
ax= sns.heatmap(flights, center=flights.loc["January", 1955])
```

```python
data= np.random.randn(50, 20) #행 :50, 열:20
ax= sns.heatmap( data, xticklabels=2, yticklabels=False)
```

```python
# 컬러바 없애기
ax= sns.heatmap(flights, cbar=False)
```

```python
ax= sns.heatmap(data, cbar=False)
```

```python
# 다른 축으로 color바를 넣기
import matplotlib.pyplot as plt
grid_kws = { "height_ratios": (.9, .05), "hspace":.3}
f, (ax, cbar_ax) = plt.subplots(2, gridspec_kw=grid_kws) #2행 위에 heatmap, 아래 cbar
ax= sns.heatmap( flights, ax=ax, 
                cbar_ax = cbar_ax, 
                cbar_kws={"orientation": "horizontal"})
```

```python
# use a mask to plot only part of a matrix (마스크를 사용)
corr= np.corrcoef(np.random.randn(10,200))
# numpy.corrcoef: return pearson product-moment correlation coefficient

mask= np.zeros_like(corr)

mask[np.triu_indices_from(mask)]=True
with sns.axes_style("white"):
    ax= sns.heatmap(corr, mask=mask, vmax=.3, square=True)
```
