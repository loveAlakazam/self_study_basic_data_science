```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

sns.set(style='ticks', color_codes=True)
```

```python
# Categorical scatterplots

tips=sns.load_dataset("tips")
sns.catplot(x='day', y='total_bill', data=tips)
```

```python
# jitter: jitter의 magnitued를 통제, magnitude를 하나로 합치는걸 못하게한다.
# jitter: 흐트러짐
sns.catplot(x='day', y='total_bill', jitter=False, data=tips);
```

```python
# swarm: 벌집모양.. swarmplot() => catplot()에서 kind가swarm
sns.catplot(x='day', y='total_bill', kind='swarm', data=tips)
```

```python
# x축: day
# y축: total_bill
# 점색깔: sex (hue)
# 그래프모양(kind): swarm
# data: 입력데이터셋(tips)
sns.catplot(x='day', y='total_bill', hue='sex', kind='swarm', data=tips)
```

```python
# x축: size
# y축: total_bill
# 그래프모양: swarm
# data: 입력데이터에서 쿼리문을 만족하는 데이터만해당.
sns.catplot(x='size', y='total_bill', kind='swarm', data=tips.query("size!=3"))
```

```python
# x축: smoker
# y축: tip
# order= 왼쪽 No, 오른쪽 Yes
# data: 입력데이터셋: tips
sns.catplot(x='smoker', y='tip', order=["No", "Yes"], data=tips)
```

```python
# x축: total_bill
# y축: day
# 점색깔(hue): time
# 그래프 모양: swarm
# 입력데이터셋 data : tips
sns.catplot(x='total_bill', y='day', hue='time', kind='swarm', data=tips)
```

```python
#Violineplots

# x축: total_bill
# y축: day
# 색깔(hue): time
# catplot의 그래프 종류: violin
# 입력데이터셋: tips
sns.catplot(x='total_bill', y='day', hue="time",
           kind='violin', data=tips)
```

```python
#bw : width값
#cut
sns.catplot(x='total_bill', y='day', hue="time",
           kind="violin", bw=.15, cut=0,
           data=tips)
```

```python
# x축: day
# y축: total_bill
# 색깔(hue)= sex
# 그래프 모양: violin
## 요일별 남성/여성 을 violin 그래프형태로 나타냄
# 입력데이터셋: tips

# 위는 split을 허용
sns.catplot(x='day', y='total_bill', hue="sex",
           kind="violin", split=True, data=tips)

#아래는 split을 비허용
sns.catplot(x='day', y='total_bill', hue="sex",
           kind="violin", split=False, data=tips)
```

```python
# x축: day
# y축: total_bill
# 색깔(hue): sex
# 그래프 모양: violin
# 안에 있는 선:stick(막대기모양으로 안에 표시)
# palette: 파스텔톤
# 입력데이터셋: tips
sns.catplot(x='day', y='total_bill', hue='sex',
           kind='violin', inner='stick', split=True,
           palette="pastel", data=tips)
```

```python
#violin모양 그래프
#inner=None: 바이올린 모양 안에 아무것도 표시 안함.
g=sns.catplot(x="day", y="total_bill", kind="violin",
             inner=None, data=tips)


# 검은 점 scatterplot 
# size: 점의 크기
# 그 모양안에 표시.
sns.swarmplot(x='day', y='total_bill', color='k', size=3,
             data=tips, ax=g.ax)
```

```python
# Bar plots
```

```python
#입력데이터를 titanic으로..
titanic= sns.load_dataset("titanic")

# x축: sex
# y축: survived
# 그래프 색깔(hue):class
# 그래프 모양: bar
# 입력데이터셋: titanic
sns.catplot(x='sex', y='survived', hue='class', kind='bar', data=titanic)
```

```python
# x축: deck
# y축: count
# palette: 색깔변경
# 입력데이터셋: data=titanic
sns.catplot(x='deck', kind='count', palette="ch:.25", data=titanic)
```

```python
# kind가 count 인 그래프는 x축이 count자체이므로 y축의 컬럼만 입력.
# y=deck
# 막대 색깔(hue): class
# palette: 파스텔
# edgecolor='.9' (선테두리 색깔값: 클수록 연함./ 작을수록 진함)
sns.catplot(y='deck', hue='class', kind='count',
           palette='pastel', edgecolor='.9',
           data=titanic)
```

```python
# kind가 count 인 그래프는 x축이 count자체이므로 y축의 컬럼만 입력.
# y=deck
# 막대 색깔(hue): class
# palette: 파스텔
# edgecolor='.6'
sns.catplot(y='deck', hue='class', kind='count',
           palette='pastel', edgecolor='.6',
           data=titanic)
```

```python
# point plots
```

```python
# x축: sex
# y축: survived
# 선색깔(hue): class
# 그래프 모양: point 
# 입력데이터셋: titanic
sns.catplot(x="sex", y="survived", hue='class', kind="point", data=titanic)
```

```python
# x축: class
# y축: survived
# 선색깔(hue): sex
# 팔렛트(지정): 남성:초록색 / 여성:마젠타
# 마커-점모양(지정): 남성: 세모/ 여성: 원
# 선스타일(지정): 남성:실선 / 여성: 점선

# 그래프모양kind: point plot
# 입력데이터셋: titanic
sns.catplot(x="class", y="survived", hue="sex",
           palette={"male":"g", "female":"m"},
           markers=["^", "o"],
           linestyles=["-", "--"],
           kind="point", data=titanic)
```
