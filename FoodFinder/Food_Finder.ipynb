# 안서동 맛집을 탐색해보자!

## 0. 웹드라이버 실행하기 & 홈페이지 열기

```python
import time
from bs4 import BeautifulSoup
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options

# 전체화면으로 키운다.
options= Options()
options.add_argument('--start-fullscreen')

driver= webdriver.Chrome('./driver/chromedriver.exe', chrome_options=options)
driver.get('https://www.diningcode.com') # 다이닝코드 홈페이지를 연다.


# (1) 광고를 닫는다.
xpath0='''//*[@id="div_popup"]/div/div[1]'''
driver.find_element_by_xpath(xpath0).click()
```

<br>

## 1. 검색하기

```python
# (2) 먼저 검색창에 "안서동" 이라고 쳐본다.
input_search= driver.find_element_by_id('txt_keyword') #검색란의 태그를 찾는다.
input_search.clear() #기존에 입력된 '음식명으로 주변 맛집을 찾아보세요.'를 지운다.
input_search.send_keys('안서동')
time.sleep(1)
```

```python
# (3) 검색창을 한번 누른다. => 아래 연관검색어가 뜬다.
xpath1='''//*[@id="txt_keyword"]'''
driver.find_element_by_xpath(xpath1).click()
time.sleep(1)
```

```python
# (4) 검색하다가 밑에 뜨는 '안서동' 자동검색을 누른다.
# 자동검색에 뜬 대상의 xpath를 얻는다.
xpath2='''//*[@id="div_search_keyup"]/ul/li/a'''
driver.find_element_by_xpath(xpath2).click()
time.sleep(1)
```

```python
# (5) '안서동' 맛집의 개수를 얻는다. => 더보기 버튼을 누른다.
from bs4 import BeautifulSoup

store_cnt_raw= driver.find_element_by_xpath('''//*[@id="lbl_normal"]''')
store_cnt= store_cnt_raw.text  # store_cnt= '130곳의 맛집' 

#'곳'의 인덱스 이전까지 문자형숫자데이터만 구한다 => int로 바꾼다.
store_cnt= int(store_cnt[: store_cnt.index('곳')]) 
print('안서동 맛집 개수: {}개'.format(store_cnt))

time.sleep(1) #1초 기다린다.

# 더보기 버튼을 누른다.
xpath3='''//*[@id="btn_normal_list"]/a''' # 더보기 버튼의 개수를 찾는다.
driver.find_element_by_xpath(xpath3).click() #더보기 버튼을 누른다.
time.sleep(1)
```

```python
import time

print('데이터를 준비합니다.')
# 지역상관없이 모두 100개의 맛집을 보여준다.
# 초기에도 10개씩만 보여주므로 9번의 버튼을 누른다.
store_name, store_menu= [],[]
for cnt in range(9): # click_cnt번 더보기 버튼을 누른다.
    driver.find_element_by_xpath('''//*[@id="div_list_more"]/a''').click()
    time.sleep(1)  
time.sleep(1)
print('데이터 준비 완료!')
```

```python
# Webdriver가 현재 실행중인 웹사이트의 소스를 가져오기.
req=driver.page_source 

# BeautifulSoup 사용.
soup= BeautifulSoup(req, 'html.parser')
#print(soup)

# 상점이름을 찾는다.
# soup.find_all('span', class_='btxt') => span태그이고, 클래스이름이 btxt이다.
# (숫자번호).' '(상점이름) 식으로 추출되므로 (상점이름)만 추출하도록 조정
store_names_raw= soup.find_all('span', class_='btxt')
print(store_names_raw)
```

```python
store_names= [ store.get_text() for store in store_names_raw]
store_names
```

```python
# get_text() => 얻은 태그에서 텍스트만을 추출
store_names= [ store.get_text()[store.get_text().index(' ')+1:] for store in store_names_raw if store.get_text()[0].isdigit()==True ]
store_names
```

## 5. 메뉴정보 수집하기

```python
# 클래스이름이 stxt인 span태그를 모두 구한다.
# 그러나 메뉴와 상관없는 불필요한 데이터가 존재한다

menu_raw=soup.find_all('span', class_='stxt')
menu_list= [ menu.get_text() for menu in menu_raw]
menu_list
```

```python
# 메뉴와 상관없는 불필요한 데이터는 삭제한다.
menu_list= [ menu.get_text() for menu in menu_raw \
            if (menu.get_text() in('광고키워드중심, 지역기반, 효율높은광고','광고네이버페이 최대 100,000P 쏜다!')) is False ]

menu_list
```

## 6. 주소정보 가져오기

```python
# class이름이 ctxt인 span 태그를 모은다.
address_raw=soup.find_all('span', class_='ctxt')

# 여러개가 중복되어있다.
address_list= [ address.get_text()[address.get_text().index('충'):] \
                for address in address_raw if '충청남도' in address.get_text() ]

print(len(address_list))
address_list
```

```python
def open_diningcode_webpage(keyword):
    #다이닝코드 홈페이지를 연다
    driver.get('https://www.diningcode.com') 
    time.sleep(1)
    
    #광고를 닫는다.
    xpath0='''//*[@id="div_popup"]/div/div[1]'''
    driver.find_element_by_xpath(xpath0).click()
    time.sleep(1)
    
    # 검색창에 keyword에 해당하는 키워드를 입력
    input_search= driver.find_element_by_id('txt_keyword') 
    input_search.clear()
    input_search.send_keys(keyword)
    time.sleep(1)
    
    # 검색창을 누른다.
    xpath1='''//*[@id="txt_keyword"]'''
    driver.find_element_by_xpath(xpath1).click()
    
    # 자동검색에 keyword와 일치한 것을 누른다=>페이지이동
    xpath2='''//*[@id="div_search_keyup"]/ul/li/a'''
    driver.find_element_by_xpath(xpath2).click()
```
